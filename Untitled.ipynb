# Data Loading & Master Dataset Creation

 ## Set Your File Paths

```python
import os
BASE_PATH = r"C:\CWH-DS\Projects\Project_10\AP_Data"

RAINFALL_PATH = os.path.join(BASE_PATH, "IMD_AP_Historical Rain Fall")
CANALS_SHP = os.path.join(BASE_PATH, "Canals", "Canals.shp")
EMBANKMENTS_SHP = os.path.join(BASE_PATH, "Embankments", "Embankments.shp")

# Verify files exist
for f in [CANALS_SHP, EMBANKMENTS_SHP]:
    status = "✅ Found" if os.path.exists(f) else "❌ NOT FOUND"
    print(f"{status}: {f}")

csv_files = [f for f in os.listdir(RAINFALL_PATH) if f.endswith('.csv')]

print(f"\n✅ Found {len(csv_files)} rainfall CSV files:")
for f in sorted(csv_files):
    print(f"   {f}")
```

 ## Load All 5 Years of Rainfall Data

```python
# Load all CSVs together — this handles all 28M+ rows
print("Loading rainfall data... (this may take 2-4 minutes)")

dfs = []
for filename in sorted(os.listdir(RAINFALL_PATH)):
    if filename.endswith('.csv'):
        filepath = os.path.join(RAINFALL_PATH, filename)
        print(f"  Reading {filename}...")
        df = pd.read_csv(filepath, 
                         dtype={
                             'district': str,
                             'mandal': str,
                             'village': str,
                             'rainfall_mm': float,
                             'centroid_lat': float,
                             'centroid_lon': float
                         },
                         parse_dates=['date'])
        dfs.append(df)

rainfall_df = pd.concat(dfs, ignore_index=True)
print(f"\n✅ Total rows loaded: {len(rainfall_df):,}")
print(f"✅ Date range: {rainfall_df['date'].min()} to {rainfall_df['date'].max()}")
print(f"✅ Districts: {rainfall_df['district'].nunique()}")
print(f"✅ Villages: {rainfall_df['village'].nunique()}")
print(rainfall_df.head(3))
```

## Load Shapefiles

```python
print("Loading shapefiles...")

canals_gdf = gpd.read_file(CANALS_SHP)
embankments_gdf = gpd.read_file(EMBANKMENTS_SHP)

print(f"✅ Canals loaded     : {len(canals_gdf)} segments")
print(f"✅ Embankments loaded: {len(embankments_gdf)} segments")
print(f"\nCanal CRS       : {canals_gdf.crs}")
print(f"Embankment CRS  : {embankments_gdf.crs}")
print(f"\nCanal columns   : {canals_gdf.columns.tolist()}")
print(f"Embank columns  : {embankments_gdf.columns.tolist()}")
```

## Get Unique Villages with Coordinates

```python
# We dont need to compute distances for 28M rows
# Just compute once per unique village centroid — much faster

print("Extracting unique village centroids...")

village_coords = rainfall_df.groupby(
    ['district', 'mandal', 'village']
).agg(
    centroid_lat=('centroid_lat', 'first'),
    centroid_lon=('centroid_lon', 'first')
).reset_index()

print(f"✅ Unique village entries: {len(village_coords):,}")
print(village_coords.head(3))
```

## Convert Villages to GeoDataFrame

```python
print("Converting villages to GeoDataFrame...")

# Create geometry points from lat/lon
geometry = [Point(row.centroid_lon, row.centroid_lat) 
            for _, row in village_coords.iterrows()]

villages_gdf = gpd.GeoDataFrame(
    village_coords, 
    geometry=geometry, 
    crs="EPSG:4326"  # WGS84 — same as shapefiles
)

print(f"✅ Villages GeoDataFrame ready: {len(villages_gdf)} rows")
print(villages_gdf.head(3))
```

## Reproject to Meters (for accurate distance in km)

```python
# EPSG:32644 is UTM Zone 44N — best for Andhra Pradesh
print("Reprojecting to UTM for accurate distance calculation...")

villages_utm    = villages_gdf.to_crs("EPSG:32644")
canals_utm      = canals_gdf.to_crs("EPSG:32644")
embankments_utm = embankments_gdf.to_crs("EPSG:32644")

print("✅ All layers reprojected to UTM Zone 44N (meters)")
```

##  Compute Distance to Nearest Embankment 

```python
# ── STEP: Find nearest embankment NAME + distance ──────────────────────────────
def nearest_embankment(village_pt, emb_gdf):
    dists = emb_gdf.geometry.distance(village_pt)
    idx = dists.idxmin()
    return emb_gdf.loc[idx, 'name'], round(dists[idx] / 1000, 3)

results = villages_utm.geometry.apply(lambda pt: nearest_embankment(pt, embankments_utm))
villages_utm['nearest_embankment_name'] = [r[0] for r in results]
villages_utm['dist_embankment_km']      = [r[1] for r in results]

print("✅ Embankment names computed")
print(villages_utm[['nearest_embankment_name', 'dist_embankment_km']].head(3))
```

#  Feature Engineering & Risk Score Calculation

##  Import & Load Master Dataset

```python
import pandas as pd
import numpy as np
import os
import warnings
warnings.filterwarnings('ignore')

# ✅ FIXED PATH
BASE_PATH = r"C:\CWH-DS\Projects\Project_10\AP_Data"

print("Loading master dataset...")
master_df = pd.read_parquet(os.path.join(BASE_PATH, "master_dataset.parquet"))

print(f"✅ Loaded {len(master_df):,} rows")
print(f"   Columns: {master_df.columns.tolist()}")
print(f"   Date range: {master_df['date'].min()} to {master_df['date'].max()}")
```

## Extract Date Features

```python
print("Extracting date features...")

master_df['date']       = pd.to_datetime(master_df['date'])
master_df['year']       = master_df['date'].dt.year
master_df['month']      = master_df['date'].dt.month
master_df['day']        = master_df['date'].dt.day
master_df['day_of_year']= master_df['date'].dt.dayofyear
master_df['week']       = master_df['date'].dt.isocalendar().week.astype(int)

# Monsoon flag (June=6 to October=10)
master_df['is_monsoon'] = master_df['month'].between(6, 10).astype(int)

print("✅ Date features extracted")
print(master_df[['date','year','month','day','is_monsoon']].head(3))
```

## Rainfall Intensity Classification

```python
# IMD Standard Rainfall Classification
def classify_rainfall(mm):
    if mm == 0:
        return 'No Rain'
    elif mm < 2.5:
        return 'Light'
    elif mm < 15.6:
        return 'Moderate'
    elif mm < 64.5:
        return 'Heavy'
    elif mm < 115.6:
        return 'Very Heavy'
    else:
        return 'Extremely Heavy'

print("Classifying rainfall intensity...")
master_df['rainfall_category'] = master_df['rainfall_mm'].apply(classify_rainfall)

print("✅ Rainfall categories:")
print(master_df['rainfall_category'].value_counts())
```

## Rolling Rainfall Features (per village)

```python
# Sort first — critical for rolling calculations
print("Sorting data... (may take 1-2 minutes)")
master_df = master_df.sort_values(['district','mandal','village','date']).reset_index(drop=True)

print("Computing rolling rainfall features...")

# Group by village for rolling stats
master_df['rainfall_3day']  = master_df.groupby(['district','mandal','village'])['rainfall_mm'].transform(
    lambda x: x.rolling(3,  min_periods=1).sum()
)
master_df['rainfall_7day']  = master_df.groupby(['district','mandal','village'])['rainfall_mm'].transform(
    lambda x: x.rolling(7,  min_periods=1).sum()
)
master_df['rainfall_30day'] = master_df.groupby(['district','mandal','village'])['rainfall_mm'].transform(
    lambda x: x.rolling(30, min_periods=1).sum()
)

# Rainfall anomaly — how much today's rain deviates from village monthly average
monthly_avg = master_df.groupby(
    ['district','mandal','village','month']
)['rainfall_mm'].transform('mean')

master_df['rainfall_anomaly'] = master_df['rainfall_mm'] - monthly_avg

print("✅ Rolling features computed")
print(master_df[['village','date','rainfall_mm','rainfall_3day','rainfall_7day','rainfall_30day','rainfall_anomaly']].head(5))
```

## Infrastructure Proximity Risk Score

```python
print("Computing infrastructure proximity risk...")

# Canal proximity risk (closer = higher risk)
def canal_proximity_score(dist_km):
    if dist_km <= 1:
        return 5   # Very High
    elif dist_km <= 3:
        return 4   # High
    elif dist_km <= 5:
        return 3   # Medium
    elif dist_km <= 10:
        return 2   # Low
    else:
        return 1   # Very Low

# Embankment proximity risk
def embankment_proximity_score(dist_km):
    if dist_km <= 1:
        return 5
    elif dist_km <= 5:
        return 4
    elif dist_km <= 10:
        return 3
    elif dist_km <= 20:
        return 2
    else:
        return 1

master_df['canal_proximity_score']      = master_df['dist_canal_km'].apply(canal_proximity_score)
master_df['embankment_proximity_score'] = master_df['dist_embankment_km'].apply(embankment_proximity_score)

print("✅ Proximity scores computed")
print(master_df[['village','dist_canal_km','canal_proximity_score',
                  'dist_embankment_km','embankment_proximity_score']].head(5))
```

##  Rainfall Risk Score 

```python
print("Computing rainfall risk score...")

def rainfall_risk_score(row):
    score = 0
    
    # Today's rainfall
    if row['rainfall_mm'] >= 115.6:
        score += 5
    elif row['rainfall_mm'] >= 64.5:
        score += 4
    elif row['rainfall_mm'] >= 15.6:
        score += 3
    elif row['rainfall_mm'] >= 2.5:
        score += 1
    
    # 3-day accumulation
    if row['rainfall_3day'] >= 200:
        score += 3
    elif row['rainfall_3day'] >= 100:
        score += 2
    elif row['rainfall_3day'] >= 50:
        score += 1
    
    # 7-day accumulation
    if row['rainfall_7day'] >= 300:
        score += 2
    elif row['rainfall_7day'] >= 150:
        score += 1
    
    # Monsoon boost
    if row['is_monsoon'] == 1:
        score += 1
        
    return score

master_df['rainfall_risk_score'] = master_df.apply(rainfall_risk_score, axis=1)

print("✅ Rainfall risk scores computed")
print(master_df['rainfall_risk_score'].describe())
```

##  Combined Risk Score & Alert Level

```python
print("Computing combined risk score...")

# Weighted combination
# Rainfall = 60%, Canal proximity = 25%, Embankment = 15%
master_df['combined_risk_score'] = (
    (master_df['rainfall_risk_score']          * 0.60) +
    (master_df['canal_proximity_score']         * 0.25) +
    (master_df['embankment_proximity_score']    * 0.15)
).round(2)

# Final Alert Level
def assign_alert(score):
    if score >= 6.0:
        return 'RED'       # Extreme Risk
    elif score >= 4.0:
        return 'ORANGE'    # High Risk
    elif score >= 2.5:
        return 'YELLOW'    # Moderate Risk
    else:
        return 'GREEN'     # Low / No Risk

master_df['alert_level'] = master_df['combined_risk_score'].apply(assign_alert)

print("✅ Alert levels assigned:")
print(master_df['alert_level'].value_counts())
print()
print(master_df[['district','mandal','village','date',
                  'rainfall_mm','combined_risk_score','alert_level']].head(10))
```

##  Verify Alert Distribution by District

```python
print("Alert distribution by district:")
alert_summary = master_df.groupby(['district','alert_level']).size().unstack(fill_value=0)
print(alert_summary)
```

# ML Model Training

## Import Libraries

```python
import pandas as pd
import numpy as np
import os
import pickle
import warnings
warnings.filterwarnings('ignore')

from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import (classification_report, confusion_matrix, 
                              accuracy_score, f1_score)
from sklearn.utils.class_weight import compute_class_weight

import matplotlib.pyplot as plt
import seaborn as sns

# ✅ FIXED PATH
BASE_PATH = r"C:\CWH-DS\Projects\Project_10\AP_Data"

print("✅ Libraries loaded")
```

## Load Featured Dataset

```python
print("Loading featured dataset...")
df = pd.read_parquet(os.path.join(BASE_PATH, "featured_dataset.parquet"))

print(f"✅ Loaded {len(df):,} rows")
print(f"   Columns: {df.columns.tolist()}")
print(f"\nAlert distribution:")
print(df['alert_level'].value_counts())
```

 ## Select Features for Model

```python
# Features the model will learn from
FEATURE_COLS = [
    'rainfall_mm',
    'rainfall_3day',
    'rainfall_7day', 
    'rainfall_30day',
    'rainfall_anomaly',
    'month',
    'day_of_year',
    'is_monsoon',
    'dist_canal_km',
    'dist_embankment_km',
    'canal_proximity_score',
    'embankment_proximity_score'
]

TARGET_COL = 'alert_level'

print("Features selected:")
for i, f in enumerate(FEATURE_COLS, 1):
    print(f"  {i:2}. {f}")
```

## Prepare Training Data (Smart Sampling)

```python
# 28M rows is too large to train directly
# We'll do STRATIFIED sampling to keep class proportions
# but also OVERSAMPLE minority classes

print("Preparing training data...")

# Step 1 — Separate by alert level
green_df    = df[df['alert_level'] == 'GREEN']
yellow_df   = df[df['alert_level'] == 'YELLOW']
orange_df   = df[df['alert_level'] == 'ORANGE']
red_df      = df[df['alert_level'] == 'RED']

print(f"GREEN  : {len(green_df):,}")
print(f"YELLOW : {len(yellow_df):,}")
print(f"ORANGE : {len(orange_df):,}")
print(f"RED    : {len(red_df):,}")

# Step 2 — Balanced sampling
# Keep all RED (rarest), sample others proportionally
n_red    = len(red_df)           # ~115k — keep all
n_orange = min(len(orange_df), n_red * 7)   # ~800k
n_yellow = min(len(yellow_df), n_red * 6)   # ~700k
n_green  = min(len(green_df),  n_red * 10)  # ~1.1M

sampled_df = pd.concat([
    red_df,
    orange_df.sample(n=n_orange, random_state=42),
    yellow_df.sample(n=n_yellow, random_state=42),
    green_df.sample(n=n_green,   random_state=42)
]).reset_index(drop=True)

print(f"\nBalanced training set: {len(sampled_df):,} rows")
print(sampled_df['alert_level'].value_counts())
```

## Handle Missing Values & Encode Target

```python
print("Preparing features & target...")

# Fill any NaN values
sampled_df[FEATURE_COLS] = sampled_df[FEATURE_COLS].fillna(0)

X = sampled_df[FEATURE_COLS]
y = sampled_df[TARGET_COL]

# Encode target labels
le = LabelEncoder()
y_encoded = le.fit_transform(y)

print(f"✅ Label encoding: {dict(zip(le.classes_, le.transform(le.classes_)))}")
print(f"   X shape: {X.shape}")
print(f"   y shape: {y_encoded.shape}")

# Save label encoder for later use in frontend
os.makedirs(os.path.join(BASE_PATH, "models"), exist_ok=True)
with open(os.path.join(BASE_PATH, "models", "label_encoder.pkl"), 'wb') as f:
    pickle.dump(le, f)

print("✅ Label encoder saved")
```

## Train/Test Split


```python
print("Splitting into train/test sets...")

X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded,
    test_size=0.20,
    random_state=42,
    stratify=y_encoded   # keeps class proportions in both splits
)

print(f"✅ Train size : {len(X_train):,}")
print(f"✅ Test size  : {len(X_test):,}")
print(f"\nTrain class distribution:")
unique, counts = np.unique(y_train, return_counts=True)
for u, c in zip(le.classes_[unique], counts):
    print(f"  {u}: {c:,}")
```

## Compute Class Weights

```python
# This tells the model to penalize mistakes on rare classes (RED) more
print("Computing class weights...")

class_weights = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(y_encoded),
    y=y_encoded
)
class_weight_dict = dict(zip(np.unique(y_encoded), class_weights))

print("Class weights:")
for idx, weight in class_weight_dict.items():
    print(f"  {le.classes_[idx]} : {weight:.4f}")
```

## Train Random Forest Model

```python
print("Training Random Forest model...")
print("(This may take 5-10 minutes...)")

rf_model = RandomForestClassifier(
    n_estimators=200,
    max_depth=20,
    min_samples_split=10,
    min_samples_leaf=5,
    class_weight=class_weight_dict,
    random_state=42,
    n_jobs=-1,        # use all CPU cores
    verbose=1
)

rf_model.fit(X_train, y_train)
print("✅ Random Forest training complete")
```

##  Evaluate Model

```python
print("Evaluating model on test set...")

y_pred = rf_model.predict(X_test)

print("\n📊 Classification Report:")
print(classification_report(y_test, y_pred, target_names=le.classes_))

print(f"\n✅ Overall Accuracy : {accuracy_score(y_test, y_pred):.4f}")
print(f"✅ Macro F1 Score   : {f1_score(y_test, y_pred, average='macro'):.4f}")
print(f"✅ Weighted F1 Score: {f1_score(y_test, y_pred, average='weighted'):.4f}")
```
